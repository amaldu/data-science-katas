### One of the most common error in DL is shape errors. 

Let's see how we can deal with those

```python
# shapes for matrix multiplication

tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[10,20],
                         [31,42],
                         [52,65]])


#torch.mm is the same as torch.matmul (it's an alias)
torch.mm(tensor_A, tensor_B)
```

We can't because of their dimensions. Let's check the shape

```python
tensor_A.shape, tensor_B.shape
```

To fix the tensor shape issues we can transpose them. 

```python
tensor_B = tensor_B.T
tensor_A.shape, tensor_B.shape
```

```python
torch.mm(tensor_A, tensor_B)
# works!
```

### Reshaping, stacking, squeezing and unsqueezing tensors

- Reshaping : reshapes an input tensor to a defined shape
- View: return a view of the input tensor of a certain shape but keep the same memory as the original tensorç
- Stacking: combine multiple tensors on top of each other (there are vertical and horiontal stacks)
- Squeeze: removes all `1` dimensions from a tensor
- Unsqueeze: adds a `1` dimension to a target tensor
- Permute: return a view of the input with dimensions permuted (swapped) in a certain way

```python
# Create a tensor

import torch
x = torch.arange(1.,10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1,7) # the dimensions have to be the same as in the original. We are trying to squeeze 9 elements into a 7 elements tensor
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(9, 2) # no because it's 18 elements!
x_reshaped, x_reshaped.shape
```

```python
# Change the view

z = x.view(1,9) # z shares the same memory place as x
z, z.shape
```

```python
# changing z vhanges x because a view of a tensor shares the same memory as the original

z[:, 0] = 5
z,x
```

```python
# stack tensors on top of each other

x_stacked = torch.stack([x,x,x,x], dim = 0) # vertically
x_stacked
```

```python
x_stacked = torch.stack([x,x,x,x], dim = 1) # horizontally
x_stacked
```

```python
x_stacked = torch.stack([x,x,x,x], dim = 2) # horizontally
x_stacked
```

```python
# squeeze: removes all single dimensions from a target tensor
print(f"Previous tensor is {x_reshaped}") # from previous cells
print(f"Previous shape is {x_reshaped.shape}") 
```

```python
print(f"Squeezed tensor is {x_reshaped.squeeze()}") # it removes a pair of []
```

```python
print(f"Previous shape is {x_reshaped.squeeze().shape}")
```

```python
# torch.unsqueeze() adds a single dimension to a target tensor at a specific dimension

print(f" The unsqueezed tensor is {x_reshaped.unsqueeze(dim=0)}")
```

```python
print(f" The unsqueezed tensor is {x_reshaped.unsqueeze(dim=1)}")
```

```python
# torch.permute() returns a view of the original input tensor with the elements swapped

tensor_original = torch.rand(size=(224,224,3))
print(f"Previous shape: {tensor_original.shape}")
```

```python
tensor_permuted = tensor_original.permute(2,0,1)
print(f"Previous shape: {tensor_permuted.shape}")
```

# Pytorh tensors and NumPy. We gonna touch the functionality to interact with each other

1. Data in NumPy want in in Pytorch tensor -> `torch.from_numpy(ndarray)`
2. Pytorch tensor -> NumPy -> `torch.Tensor.numpy()`

```python
import numpy as np
import torch
array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array)
array, tensor

# important to remember! When coming from numpy, the float64 is by dafault. When the dafault precision for torch tensors is float32
```

```python
new_tensor = torch.from_numpy(array).type(torch.float32)
new_tensor.dtype
```

```python
# change the value of an array, what will this do to the tensor?

array = array +1 

array, tensor

# it changes the value of the array but not the tensor
```

```python
# tensor to numpy array

tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
numpy_tensor

# it's float32 by default. Be careful beause an error of data types was a common error
```

```python
# change tensor what happens to numpy_tensor?

another_new_tensor = tensor + 1
another_new_tensor, numpy_tensor

# same as before, it does not change. They don't change memory
```

# Reproducibility

Trying to take random out of random. We use random seed to flavour the randomness



```python
import torch

# create random tensors


random_tensor_A = torch.rand(3,5)
random_tensor_B = torch.rand(3,5)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# let's make some random but reproducible tensors

import torch
# set the random seed

RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,5)
torch.manual_seed(RANDOM_SEED) # we need to set the seed to make it work
random_tensor_D = torch.rand(3,5)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```

To know more about it -> https://docs.pytorch.org/docs/stable/notes/randomness.html
