### you are given the table named financial_transactions containing information on user transactions for a particular business using Stripe. Write a query to obtain the accountâ€™s rolling 7-day earnings.

```python
%%sql

SELECT * FROM financial_transactions LIMIT 10;
```

```python
%%sql

SELECT
    ft1.transaction_id,
    ft1.user_id,
    ft1.amount,
    ft1.transaction_date,
    (
        SELECT SUM(ft2.amount)
        FROM financial_transactions ft2
        WHERE ft2.transaction_date BETWEEN ft1.transaction_date - INTERVAL '6 days'
                                      AND ft1.transaction_date
    ) AS rolling_7day_earnings
FROM financial_transactions ft1
WHERE ft1.transaction_date >= NOW() - INTERVAL '7 days'
ORDER BY ft1.transaction_date;



```
