### you are given the table named measurements containing measurement values obtained from a sensor over several days. Measurements are taken several times within a given day. Write a query to obtain the sum of the odd-numbered measurements and the sum of the even-numbered measurements by date.

```python
%%sql

SELECT * FROM measurements LIMIT 10;
```

```python
%%sql

SELECT
    DATE(measurement_time) AS measurement_date,
    SUM(CASE WHEN measurement_id % 2 = 1 THEN measurement_value ELSE 0 END) AS odd_sum,
    SUM(CASE WHEN measurement_id % 2 = 0 THEN measurement_value ELSE 0 END) AS even_sum
FROM measurements
GROUP BY DATE(measurement_time)
ORDER BY measurement_date;
```
