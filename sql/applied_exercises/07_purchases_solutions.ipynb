### You are given the table named purchases. Write a query to obtain the number of people who purchased at least one or more of the same product on multiple days

```python
%%sql

SELECT * FROM purchases LIMIT 10;
```

Ok so we need to check how many users buy the same product and detect repetitions of the product in different days (DISTINCT)


```python
%%sql

SELECT DISTINCT user_id
FROM purchases
GROUP BY user_id, product_id
HAVING COUNT(DISTINCT DATE(purchase_time)) >= 2; 
```

### Now write a query that shows the user_id, purchase_id and purchase_day for these cases

```python
%%sql
SELECT DISTINCT 
    user_id,
    product_id,
    DATE(purchase_time) AS purchase_day
FROM purchases
WHERE user_id IN (
    SELECT user_id
    FROM purchases
    GROUP BY user_id, product_id
    HAVING COUNT(DISTINCT DATE(purchase_time)) >= 2
)
ORDER BY user_id, product_id, purchase_day;
```
