### You are given the table total_trans for spending activity by product type. Write a query to calculate the cumulative spend so far by date for each product over time in chronological order

```python
%%sql
SELECT * FROM total_trans LIMIT 10
```

```python
%%sql
SELECT product_id, spend, SUM(spend) OVER (PARTITION BY product_id ORDER BY product_id, trans_date ASC) AS cummulative_spend
FROM total_trans
```
