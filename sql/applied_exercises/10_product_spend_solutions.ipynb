### You re given the table product_spend. Identify the top three highest-grossing items within each category in 2025.

```python
%%sql

SELECT * FROM product_spend LIMIT 10;
```

```python
%%sql

WITH category_product_revenue AS (
    SELECT
        category_id,
        product_id,
        SUM(spend) AS total_revenue
    FROM product_spend
    WHERE EXTRACT(YEAR FROM transaction_date) = 2025
    GROUP BY category_id, product_id
)
SELECT
    category_id,
    product_id,
    total_revenue
FROM (
    SELECT
        category_id,
        product_id,
        total_revenue,
        ROW_NUMBER() OVER (
            PARTITION BY category_id
            ORDER BY total_revenue DESC
        ) AS revenue_rank
    FROM category_product_revenue
) ranked
WHERE revenue_rank = 1
ORDER BY category_id;


```
