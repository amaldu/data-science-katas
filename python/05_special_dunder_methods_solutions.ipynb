## 2. add  `__repr__()` method that returns a string that we can use to recreate the object

Instantiate the Employee class twice under the names "emp_1" and "emp_2" 

Print emp_1. What happens?

```python
class Employee:
    raise_amt = 1.05
    def __init__(self, first, last, salary):
        self.first = first
        self.last = last
        self.email = f"{first.lower()}.{last.lower()}@example.com"
        self.salary = salary

    def full_name(self):
        return f"{self.first} {self.last}"
    
    def apply_raise(self):
        self.salary = int(self.salary * self.raise_amt)
        return self.salary
    
    def __repr__(self):
        return "Employee('{}', '{}', '{}')".format(self.first, self.last, self.salary)
    
emp_1 = Employee("Lucas", "Mann", 100000)
emp_2 = Employee("Sophia", "Williams", 80000)

print(emp_1)
print(repr(emp_1))
print(emp_1.__repr__())

# It creates an employee representation of what it shouls be passed to the class when we instantiate it
```

## 3. `__str__()` method:

it's very similar to __repr__() but it's meant to be a readable representation of an object and is used as a display to the end user in a pretty way.

Rewrite the Employee class as in the last question and add __str__ method 

```python
class Employee:
    raise_amt = 1.05
    def __init__(self, first, last, salary):
        self.first = first
        self.last = last
        self.email = f"{first.lower()}.{last.lower()}@example.com"
        self.salary = salary

    def full_name(self):
        return f"{self.first} {self.last}"
    
    def apply_raise(self):
        self.salary = int(self.salary * self.raise_amt)
        return self.salary
    
    def __repr__(self):
        return "Employee('{}', '{}', '{}')".format(self.first, self.last, self.salary)
    
    def __str__(self):
        return '{} - {}'.format(self.full_name(), self.email)
    
emp_1 = Employee("Lucas", "Mann", 100000)
emp_2 = Employee("Sophia", "Williams", 80000)

print(emp_1)
print(str(emp_1))
print(emp_1.__str__())
```

## 5. `__len__()` method:

This method defines what is the length of an bject when we use the function len() on it

```python
class Employee:
    raise_amt = 1.05
    def __init__(self, first, last, salary):
        self.first = first
        self.last = last
        self.email = f"{first.lower()}.{last.lower()}@example.com"
        self.salary = salary

    def full_name(self):
        return f"{self.first} {self.last}"
    
    def apply_raise(self):
        self.salary = int(self.salary * self.raise_amt)
        return self.salary
    
    def __repr__(self):
        return "Employee('{}', '{}', '{}')".format(self.first, self.last, self.salary)
    
    def __str__(self):
        return '{} - {}'.format(self.full_name(), self.email)
    
    def __add__(self, other):
        return self.salary + other.salary
    
    def __len__(self):
        return len(self.full_name())
    
emp_1 = Employee("Lucas", "Mann", 100000)
emp_2 = Employee("Sophia", "Williams", 80000)

print(emp_1.__len__())
```
